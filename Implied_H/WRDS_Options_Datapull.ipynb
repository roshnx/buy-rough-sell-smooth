````python
from datetime import datetime

start_date = datetime.strptime("20100101", "%Y%m%d")  #  start date
end_date = datetime.strptime("20230831", "%Y%m%d")    #  end date

tables = db.list_tables(library="taqmsec")
filtered_tables = []


for table in tables:
    parts = table.split('_')
    
    if len(parts) == 3 and parts[0] == "complete" and parts[1] == "nbbo":
        table_date_str = parts[2]
        
        try:
            table_date = datetime.strptime(table_date_str, "%Y%m%d")
            if start_date <= table_date <= end_date:
                filtered_tables.append(table)
        except ValueError:
            continue  # Ignore entries that don't have a valid date
    else:
        print(f"Skipping table {table}")

print(filtered_tables[-5:]) 
````

````python
def extract_option_metrics(tickers: list[str], year: int):
    tickers_str = "(" + ", ".join(["'{}'".format(t) for t in tickers]) + ")"

    sql = f"""
    SELECT 
        v.date,
        v.days,
        v.impl_volatility,
        v.impl_strike,
        v.cp_flag,
        (SELECT n.ticker 
         FROM optionm_all.optionmnames n 
         WHERE n.secid = v.secid 
         LIMIT 1) AS ticker
    FROM optionm_all.vsurfd{year} v
    WHERE EXISTS (
        SELECT 1
        FROM optionm_all.optionmnames n 
        WHERE n.secid = v.secid 
          AND n.ticker IN {tickers_str}
    )
    LIMIT 10
    """

    return db.raw_sql(sql)

tickers = ["NVDA"]
year = 2021
df = extract_option_metrics(tickers, year)
df.head()
````
